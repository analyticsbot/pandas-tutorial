## What is pandas and why to use it?

A Python package providing fast, flexible, and expressive data structures designed to make working with relational or labeled data both easy and intuitive.

- Works with spreadsheet, tables, databases, ordered, unordered
- Two basic structures - Series & DataFrame; analogous to R dataframe
- Works & handles missing data
- Multiple operations on dataframe possible such as inserting, deletion, adding rows/ columns
- Powerful groupby functionality for aggregating and transforming data
- Slicing, indexing, subsetting, reshaping data with one line code
- Hierarchial axis
- Pandas is fast (used Cython)
